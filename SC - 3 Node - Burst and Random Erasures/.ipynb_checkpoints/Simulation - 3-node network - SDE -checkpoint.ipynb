```python
import math
import numpy as np
import pandas  as pd 
import matplotlib as plot 
```

```python
def del_spec(a,b,N,k,n):
    ds = np.zeros(n)
    for i in range(0,n):
        ds[i] = N-b*(math.floor(i/a))-(i%a)
    return ds
```

## Program for finding test cases satisfying the constraints

```python
cnt1 = 0
cnt2= 0
CNT = 0

for T in range(4,40):
    for a1 in range(1,T):
        for a2 in range(1,T):
            for b1 in range(a1+1,T):
                 for b2 in range(a2+1,T):
                        if(a1==a2  and T >= b2 + b1):
                            flag_dc = 0
                            flag_check = 0
                            max_del = -1
                            max_del_ind = -1
                            k = max(a1,a2)*min((T-b2-a1+1)/b1,(T-b1-a2+1)/b2)
                            if(k.is_integer()):
                                k = int(k)
                                CNT+=1
                                N1 = (k+a1-1)+(b1-a1)*math.floor((k+a1-1)/a1)+1
                                N2 = (k+a2-1)+(b2-a2)*math.floor((k+a2-1)/a2)+1
                                if (N1<=T-b2+1 and N2<=T-b1+1):
                                    cnt1+= 1
                                    ds1=del_spec(a1,b1,N1-1,k,k+a1)
                                    ds2=del_spec(a2,b2,N2-1,k,k+a2)
                                    for i in range(0,k):
                                        if(ds1[i]+ ds2[k-i-1]>max_del):
                                            max_del_ind = i 
                                            max_del = ds1[i]+ ds2[k-i-1]
                                        if (ds1[i]+ ds2[k-i-1])>T:
                                            flag_dc = 1

                                    if(flag_dc == 1):
                                        cnt2 +=1
                                        print(k,a1,a2,T,b1,b2,N1,N2)
                                        print(ds1)
                                        print(ds2)
```

## Program for finding max rate using SDE

```python
data = []
data_m = []

cnt = 0 
CNT = 0
CNT2 = 0

for T in range(4,20):
    for a1 in range(1,T):
        for a2 in range(1,T):
            for b1 in range(a1+1,T):
                 for b2 in range(a2+1,T):
                    if(T >= b1 + b2):
                        R1_opt= (T-b2-a1+1)/(T-b2-a1+1+b1)
                        R2_opt= (T-b1-a2+1)/(T-b1-a2+1+b2)
                        R_opt = min(R1_opt,R2_opt)
                        R_max = -1
                        k_max = -1
                        cnt+=1 
                        for k in range(1,min((T-b2+1-a1),(T-b1+1-a2))+1):
                            R=-1 
                            R1 = -1 
                            R2 = -1 
                            flag_dc = 0
                            N1 = (k+a1-1)+(b1-a1)*math.floor((k+a1-1)/a1)+1
                            N2 = (k+a2-1)+(b2-a2)*math.floor((k+a2-1)/a2)+1
                            if(N1<=T-b2+1) and (N2<=T-b1+1):
                                CNT+=1
                                ds1=del_spec(a1,b1,N1-1,k,k+a1)
                                ds2=del_spec(a2,b2,N2-1,k,k+a2)
                                for i in range(0,k):
                                    if (ds1[i]+ ds2[k-i-1])>T:
                                        flag_dc = 1
                                if(flag_dc == 0):
                                    CNT2+=1
                                    R1 = k/(k+a1)
                                    R2 = k/(k+a2)
                                    R = min(R1,R2)
                                    if(R>R_max):
                                        k_max = k 
                                        R_max = R
                                data.append((T,a1,b1,N1,a2,b2,N2,k,R1,R1_opt,R2,R2_opt,R,R_opt))                               
                        data_m.append((T,a1,b1,a2,b2,k_max,R_max,R_opt,R_max==R_opt))                        
```

## Maximum Rate using SDE for a given set of (T,a1,b1,a2,b2)

```python
T = 10
a1 = 1
b1 = 2 
a2 = 2
b2 = 5

cnt = 0 
CNT = 0
CNT2 = 0

if(T>= b1 + b2):
    R1_opt= (T-b2-a1+1)/(T-b2-a1+1+b1)
    R2_opt= (T-b1-a2+1)/(T-b1-a2+1+b2)
    R_opt = min(R1_opt,R2_opt)
    R_max = -1
    k_max = -1
    cnt+=1 
    for k in range(1,min((T-b2+1-a1),(T-b1+1-a2))+1):
        R=-1 
        R1 = -1 
        R2 = -1 
        flag_dc = 0
        N1 = (k+a1-1)+(b1-a1)*math.floor((k+a1-1)/a1)+1
        N2 = (k+a2-1)+(b2-a2)*math.floor((k+a2-1)/a2)+1
        if(N1<=T-b2+1) and (N2<=T-b1+1):
            CNT+=1
            ds1=del_spec(a1,b1,N1-1,k,k+a1)
            ds2=del_spec(a2,b2,N2-1,k,k+a2)
            
            for i in range(0,k):
                if (ds1[i]+ ds2[k-i-1])>T:
                    flag_dc = 1
            if(flag_dc == 0):
                CNT2+=1
                R1 = k/(k+a1)
                R2 = k/(k+a2)
                R = min(R1,R2)
                if(R>R_max):
                    k_max = k 
                    R_max = R
            print(T,a1,b1,N1,a2,b2,N2,k,R1,R1_opt,R2,R2_opt,R,R_opt)
            print(ds1,ds2)
```

```python
cnt = 0 
CNT = 0
CNT2 = 0
data = []
data_m = []
for T in range(4,20):
    for a1 in range(1,T):
        for a2 in range(1,T):
            for b1 in range(a1+1,T):
                 for b2 in range(a2+1,T):
                    if(a1!=a2 and b1-a1==b2-a2 and T >= b1 + b2):
                        R1_opt= (T-b2-a1+1)/(T-b2-a1+1+b1)
                        R2_opt= (T-b1-a2+1)/(T-b1-a2+1+b2)
                        R_opt = min(R1_opt,R2_opt)
                        R_max = -1
                        k_max = -1
                        cnt+=1 
                        m1 = math.floor((T-b2+1)/b1)
                        d1 = (T-b2+1)%b1 
                        d1_dash = min(a1,d1)
                        k1 = (m1-1)*a1 + d1_dash
                        m2 = math.floor((T-b1+1)/b2)
                        d2 = (T-b1+1)%b2 
                        d2_dash = min(a2,d2)
                        k2 = (m2-1)*a2 + d2_dash   
                        for k in range(1,min((T-b2+1-a1),(T-b1+1-a2))+1):
                            R=-1 
                            R1 = -1 
                            R2 = -1 
                            flag_dc = 0
                            N1 = (k+a1-1)+(b1-a1)*math.floor((k+a1-1)/a1)+1
                            N2 = (k+a2-1)+(b2-a2)*math.floor((k+a2-1)/a2)+1
                            if(N1<=T-b2+1) and (N2<=T-b1+1):
                                CNT+=1
                                ds1=del_spec(a1,b1,N1-1,k,k+a1)
                                ds2=del_spec(a2,b2,N2-1,k,k+a2)
                                for i in range(0,k):
                                    if(ds1[i]+ ds2[k-i-1]>max_del):
                                            max_del_ind = i 
                                            max_del = ds1[i]+ ds2[k-i-1]
                                    if (ds1[i]+ ds2[k-i-1])>T:
                                        flag_dc = 1
                                if(flag_dc == 0):
                                    CNT2+=1
                                    R1 = k/(k+a1)
                                    R2 = k/(k+a2)
                                    R = min(R1,R2)
                                    if(R>R_max):
                                        k_max = k 
                                        R_max = R
                                data.append((T,a1,b1,N1,a2,b2,N2,k,R1,R1_opt,R2,R2_opt,R,R_opt))                               
                        data_m.append((T,a1,b1,a2,b2,k_max,k1,m1,d1,k2,m2,d2,R_max,R_opt))                        
```
