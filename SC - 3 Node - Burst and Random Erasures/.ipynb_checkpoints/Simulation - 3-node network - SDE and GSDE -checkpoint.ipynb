```python
def lcm (a,b):
    return int((a*b)/math.gcd(a,b))

def del_spec(a,b,N,k,n):
    ds = np.zeros(n)
    for i in range(0,n):
        ds[i] = N-b*(math.floor(i/a))-(i%a)
    return ds
```

# GSDE 

```python
def sgn(x):
    if x>0: return 1
    elif x==0: return 1
    else: return -1

def disp_vect(N,a,b,g,t):
    dv = [0]*(N)    
    for i in range(0,N):
        if((i+1)%b == 1):
            dv[i] = g
        else:
            dv[i] = t
    return dv
    
def ds_value (t,N,k,dv):
    temp = N - 1
    for j in range(0,N):
        sum_temp = 0
        for l in range (0,j+1):
            sum_temp = sum_temp + dv[l]
#         print(t,j,sum_temp)
        temp = temp-(1+sgn(t-sum_temp))/2
    return temp 
    
def del_spec_GSDE(N,k,dv):
    ds = [0]*(k)   
    for t in range(0,k):
        ds[t] = ds_value(t,N,k,dv)
    return ds
```

```python
T = 21
a1 = 6 
b1 = 7 
a2 = 6
b2 = 7
#                   Delay Constraint 
T1 = T-b2
T2 = T-b1

#                   Dispersion Span          
N1 = T1+1
N2 = T2+1

#                   Ti+1 = mi*bi+di
m1 = math.floor((T1+1)/b1)
d1 = int((T1+1)%b1)
m2 = math.floor((T2+1)/b2)
d2 = int((T2+1)%b2) 

#                   Parameters for dispersion vector                     
t1 = int(lcm(b1-a1,m1)/(b1-a1))
g1 = int(t1*(1+(b1-a1)/m1))
t2 = int(lcm(b2-a2,m2)/(b2-a2))
g2 = int(t2*(1+(b2-a2)/m2))

#                   Code length and redundancy                     
n1 = int(t1*(m1*b1+(m1+1)*(b1-a1)/m1+d1))
r1 = int(t1*(b1+(b1-a1)/m1))
n2 = int(t2*(m2*b2+(m2+1)*(b2-a2)/m2+d2))
r2 = int(t2*(b2+(b2-a2)/m2))

k1=n1-r1
k2=n2-r2

#                   Dispersion vector  
dv1 = disp_vect(N1,a1,b1,g1,t1)
dv2 = disp_vect(N2,a2,b2,g2,t2)

#                   Delay Spectrum           
ds1 = del_spec_GSDE(N1,n1-r1,dv1)
ds2 = del_spec_GSDE(N2,n2-r2,dv2)
R_GSDE = min(k1/n1,k2/n2)
```
