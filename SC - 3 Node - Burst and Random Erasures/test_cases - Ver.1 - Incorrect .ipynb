```python
import math
import numpy as np
import pandas  as pd 
import matplotlib as plot 
```

```python
def del_spec(a,b,N,k,n):
    ds = np.zeros(n)
    for i in range(0,n):
        ds[i] = N-b*(math.floor(i/a))-(i%a)
#         print(ds[i])
        N = ds[0]-ds[n-1]
    return ds
```

```python
# Program for finding test cases satisfying the constraints
cnt = 0
CNT = 0
for T in range(4,40):
    for a1 in range(1,T):
        for a2 in range(1,T):
            for b1 in range(a1+1,T):
                 for b2 in range(a2+1,T):
                        if(a1==a2 and b1!=b2):
                            flag_dc = 0
                            flag_dsc = 0
                            if(T - b2 > b1 and T - b1 > b2):
                                k = max(a1,a2)*min((T-b2-a1+1)/b1,(T-b1-a2+1)/b2)
                                if(k.is_integer()):
                                    k = int(k)
                                    CNT +=1
                                    ds1=del_spec(a1,b1,T-b2,k,k+a1)
                                    ds2=del_spec(a2,b2,T-b1,k,k+a2)

#----------------------------------Delay Constraint----------------------------------------------------- 
                                    for i in range(0,k):
                                        if (ds1[i]+ ds2[k-i-1])>T:
#                                         if (ds1[a1-1]+ ds2[k-a1])>T:
                                            flag_dc = 1
# ------------------------------------------------------------------------------------------------------                                    
                                    if(flag_dc ==0 and ds2[k+a2-1]>=0 and ds1[k+a1-1]>=0):
                                        cnt +=1
                                        print(k,T,a1,b1,a2,b2)
                                        print(ds1)
                                        print(ds2)
```

## Rate Calculator

```python
data = []
for T in range(4,20):
    for a1 in range(1,T):
        for a2 in range(1,T):
            for b1 in range(a1+1,T):
                 for b2 in range(a2+1,T):
                    if(a1==a2 and b1!=b2 and T - b2 > b1 and T - b1 > b2):
                        R_opt = min((T-b2-a1+1)/(T-b2-a1+1+b1),(T-b1-a2+1)/(T-b1-a2+1+b2))
                        for k in range(1,T):
                            R=-1 
                            flag_dc = 0
                            flag_dsc = 0
                            ds1=del_spec(a1,b1,T-b2,k,k+a1)
                            ds2=del_spec(a2,b2,T-b1,k,k+a2)
                            for i in range(0,k):
                                if (ds1[i]+ ds2[k-i-1])>T:
                                    flag_dc = 1
                            if(ds2[k+a2-1]>=0 and ds1[k+a1-1]>=0):
                                flag_dsc = 1 
                            if(flag_dc == 0 and flag_dsc==1):
                                R = k/(k+max(a1,a2))
                            temp1 = math.floor((T-b2-1)/b1)*a1 + min(a1,(T-b2+1)%b1)-a1
                            temp2 = math.floor((T-b1-1)/b2)*a2 + min(a2,(T-b1+1)%b2)-a2
                            data.append((T,a1,a2,b1,b2,k,temp1,temp2,R,R_opt))
```

```python
df = pd.DataFrame(data, columns = ['T', 'a1',"a2","b1","b2","k","temp1","temp2","R","R_opt"])
```
