```python

cnt = 0

for T in range(4,40):
    for a1 in range(1,T):
        for a2 in range(a1+1,T):
            for b1 in range(a1+1,T):
                 for b2 in range(a2+1,T):
                        if(T - b2 > b1 and T - b1 > b2):
                            k = max(a1,a2)*min((T-b2-a1+1)/b1,(T-b1-a2+1)/b2)
                            if math.floor(k) == k:
                                if (k*(T-b2-a1+1+b1)==(k+a1)*(T-b2-a1+1) and (k*(T-b1-a2+1+b2)==(k+a2)*(T-b1-a2+1))):
                                    cnt +=1
                                    print(k,T,a1,b1,a2,b2)
```
